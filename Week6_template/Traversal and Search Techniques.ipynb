## Tree Traversal

Three different type of tree traversal 

#### Code

```python

class TreeNode:
    def __init__(self,lc,rc,val):
        self.lchild = lc
        self.rchild = rc
        self.data = val
        
def InOrder(u):
    if u is None:
        return
    InOrder(u.lchild)
    print(u.data)
    InOrder(u.rchild)
        
        
def PreOrder(u):
    if u is None:
        return
    print(u.data)
    PreOrder(u.lchild)
    PreOrder(u.rchild)
        
        
def PostOrder(u):
    if u is None:
        return
    PostOrder(u.lchild)
    PostOrder(u.rchild)
    print(u.data)

        
t = []
for i in range(9):
    t.append(TreeNode(None,None,chr(ord('A')+i)))

t[0].lchild = t[1]
t[0].rchild = t[2]

t[1].lchild = t[3]
t[1].rchild = t[4]

t[3].lchild = t[5]
t[3].rchild = t[6]

t[6].lchild = t[7]
t[6].rchild = t[8]

InOrder(t[0])
print()
PreOrder(t[0])
print()
PostOrder(t[0])
```

## BFS and DFS

#### Code

```python
import queue
n = 8
adj = [[] for i in range(8)]

adj[0] = [1,2]
adj[1] = [0,3,4]
adj[2] = [0,5,6]
adj[3] = [1,7]
adj[4] = [1,7]
adj[5] = [2,7]
adj[6] = [2,7]
adj[7] = [3,4,5,6]



vis = [0 for i in range(n)]

def bfs(st):
    global vis
    q = queue.Queue()
    vis[st] = 1
    q.put(st)
    while not q.empty():
        u = q.get()
        for v in adj[u]:
            if vis[v]==0:
                vis[v] = 1
                q.put(v)


vis = [0 for i in range(n)]

def dfs(u):
    global vis
    
    vis[u] = 1
    for v in adj[u]:
        if vis[v]==0:
            dfs(v)
```

## Articulation Point

#### Code

```python

n = 10
adj = [[] for i in range(n)]



dfsNumCounter = 0
dfsRoot = 0
dfsRootCh = 0
L = [-1 for i in range(n)]
dfn = [-1 for i in range(n)]
art_point = [0 for i in range(n)]

def Art(u,parent):
    
    global L,dfn,art_point,dfsRootCh,dfsNumCounter
    
    L[u] = dfsNumCounter
    dfn[u] = dfsNumCounter 
    dfsNumCounter+=1
    for v in adj[u]:
        if dfn[v]==-1:
            Art(v,u)
            if u == dfsRoot: dfsRootCh+=1
            if L[v]>=dfn[u]:
                art_point[u] = 1
            L[u] = min(L[u],L[v])
        else:
            if v!=parent:
                L[u] = min(L[u],dfn[v])
                

dfsRoot = 0
dfsRootCh = 0
Art(dfsRoot,-1)
if dfsRootCh>1: art_point[0] = 1

for i in range(n):
    if art_point[i]==1: print(i)
    

```
