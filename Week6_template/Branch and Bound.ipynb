## 0/1 Knapsack Problem

#### Code

```python
import queue

def GreedyKnapsack(m,n,p,w):
    x = []
    for i in range(n): x.append((p[i]/w[i],i))
    x.sort()
    x.reverse()
    u = m
    val = 0
    y = [0 for i in range(n)] 
    for i in range(n):
        if u==0: break
        p = x[i][1]
        t = min(u,w[p])
        val+=t*x[i][0]
        u-=t
        if t== w[p]: y[p] = 1
        else: 
            if t>0: y[p] = 0.5
    return [val,y]


def Knapsack(m,n,p,w):
    
    q = queue.Queue()
    lb = 0
    q.put([])
    
    while not q.empty():
        u = q.get()
        #print(u)
        cn = 0
        cm = m
        cval = 0
        for i in range(len(u)):
            if u[i]==1:
                cm-=w[i]
                cval+=p[i]
        
        if cm<0: continue
        if len(u)==n:
            lb = max(lb,cval)
            continue
        
        cn = n-len(u)
        cp = []
        cw = []
        for i in range(len(u),n):
            cp.append(p[i])
            cw.append(w[i])
        
        re = GreedyKnapsack(cm,cn,cp,cw)
        
        if re[0]+cval<lb: continue
        
        cu = u.copy()
        cu.extend(re[1])
        
        flag = 0
        for i in range(n):
            if cu[i]!=0 and cu[i]!=1:
                flag = 1
        
        if flag==1:
            lc = u.copy()
            lc.append(0)
            rc = u.copy()
            rc.append(1)
            q.put(lc)
            q.put(rc)
        else:
            lb = max(lb,cval+re[0])
        
    return lb
    
print(Knapsack(14,4, [10,10,12,18], [2,4,6,9]))

        
```
