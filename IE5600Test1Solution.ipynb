```python
# P161 v1
test = int(input())


def print_cadies(a):
    if len(a) > 0: 
        print(' '.join(map(str, a)))
    else: 
        print(-1)


for _ in range(test):
    n = int(input())
    a = list(map(int, input().split()))
    m = int(input())
    p = list(map(int, input().split()))

    a.sort()
    temp = [] #use temp to store the left candies in each iteration temporally
    for j in range(m):
        temp = []
        b = [] #use b to store the candies of kid j
        for i in range(len(a)):
            if a[i] % p[j] == 0: 
                b.append(a[i])
            else: temp.append(a[i])
        print_cadies(b)
        a = temp

    print_cadies(a) # left candies
```

```python
# P161 v3
test = int(input())
if 1 <= test <= 100:
    for _ in range(test):
        n = int(input())
        A = [int(x) for x in input().split()]
        m = int(input())
        if m < 1 or m > 5:
            exit()
        B = [int(x) for x in input().split()]

        for b in B:
            candies = []
            for a in A.copy(): # use A.copy() to create a new array copy, so A.remove(a) inside the for loop has no effect 
                if a % b == 0:
                    candies.append(a) #use candies to store the candies of kid b
                    A.remove(a)
                if not candies:
                    print(-1)
                else:
                    print(" ".join([str(x) for x in sorted(candies)]))
        if not A:
            print(-1)
        else:
            print(" ".join([str(x) for x in sorted(A)]))
```

```python
#P163 v12
test = int(input())
for _ in range(test):
    l,x,T = map(int,input().split())
    #Implement your code here
    velocity = 1 # use velocity to indicate the direction, 1: right, -1: left
    position = x
    for i in range(T):
        position += velocity # simulate the bouncing process
        if position == l or position == 0:
            velocity *= -1
    #Output Solution
    print(position) 
```

```python
#P165  v1
import sys
t = int(input())
for _ in range(t):
    n = int(input())
    x = list(map(int,input().split()))
    x.sort()
    ans = sys.maxsize
    for i in range(1,n):
        ans = min(ans,x[i]-x[i-1])
    print(ans)
```

```python
#P165  v2
test = int(input())
for _ in range(test):
    n = int(input())
    x = [int(i) for i in input().split()]
    #Implement your Code Here
    mini_distance = 10**6
    for i in range(n):
        for j in range(i+1, n): #compare pairs one by one 
            dist = x[i] - x[j]
            if abs(dist) < mini_distance:
                mini_distance = abs(dist)
    #Output Solution
    print(mini_distance)
```
